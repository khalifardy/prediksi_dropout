```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
datas = pd.read_csv('/home/khalifardy/Dokumen/ruang_kerja/code/kuliah/ML/Assignment_CLO1_project_base/prediksi_dropout/pre_processing_data/data_eda_norm.csv')
datas
```

```python
X = datas[datas.columns[:-1]]
Y = datas.Target
```

```python
X
```

```python
Y
```

```python
label_mapping = {
    "Dropout":1,
    "Graduate":0,
    "Enrolled":0
}

Y = [label_mapping[i] for i in Y]
```

```python
Y
```

```python
#membagi dataset 
X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size=0.2, random_state=22)
```

```python
#one versus all
model = svm.SVC(kernel='rbf')
```

```python
#train model
model.fit(X_train,y_train)
```

```python
#prediksi 
y_pred = model.predict(X_test)
```

```python
#Evaluasi model
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
cm = confusion_matrix(y_test, y_pred) 
print(cm)
```
